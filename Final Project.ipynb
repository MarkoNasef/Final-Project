```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns  
```

```python
# Q1: Open Your CSV file and print the first 10 rows in a good format

data = pd.read_csv("tmdb_5000_movies.csv")                   # will read the file
data.head(10)                                               # Printing the first 10 rows
```

```python
# Q2-P1: search for NAN Value 

#True â†’ Missing (NaN) value
#False â†’ Non-missing value
data.isna().sum() #sum will get data per column without it will give me the whole sheet 
```

```python
# Q2-P2: Replace Missing Values with the word Missing

data.replace(np.nan , "Missing" , inplace = True)     
```

```python
# Q3: Discover the types of each columns and modify them if needed

data.info ()                                               #summary of a DataFrame
```

```python
data.columns                                          #Getting columns names
```

```python
# Convert 'release_date' to datetime format

data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')

# Convert runtime to numeric (non-numeric values will become NaN)
data['runtime'] = pd.to_numeric(data['runtime'], errors='coerce')
```

```python
# Q4: Statistical Overview

print("\nDescriptive statistics:")
print(data.describe())
```

```python
# Q5: Calculate the mean rate for [1999, 1980, 2004, 2017] the movies grouped by years

# Extract the release year from the datetime column
data['release_year'] = data['release_date'].dt.year

# Define the years of interest
target_years = [1999, 1980, 2004, 2017]

# Filter the dataset for those years
filtered_data = data[data['release_year'].isin(target_years)]

# Group by release_year and calculate the mean of vote_average
mean_vote_by_year = filtered_data.groupby('release_year')['vote_average'].mean()

# Display the result
print("Mean vote average for the selected years:")
print(mean_vote_by_year)
```

```python
# Q6: Rearange the dataframe based on revenue - budget values 

# Calculate 'profit' as revenue - budget
data['profit'] = data['revenue'] - data['budget']

# Sort the DataFrame by 'profit' in descending order
data_sorted_by_profit = data.sort_values(by='profit', ascending=False)
data_sorted_by_profit = data_sorted_by_profit.reset_index(drop=True)  #to remove the original file index from the results 

# Display the top 5 rows to verify
print(data_sorted_by_profit[['title', 'budget', 'revenue', 'profit']].head())
```

```python
# Q7: Allocate the year with the largest movies released in ['Action', 'Romance']

#sheet already loaded 
#dates already converted to datatime 

import ast 

# Convert genre strings to lists of dictionaries
data['genres'] = data['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Extract only the genre names into a list
data['genre_names'] = data['genres'].apply(lambda x: [d['name'] for d in x])

# Filter for movies that have 'Action' or 'Romance' in genre_names
action_romance = data[data['genre_names'].apply(lambda genres: 'Action' in genres or 'Romance' in genres)]

# Count movies per release_year
year_counts = action_romance['release_year'].value_counts()

# Get the year with the maximum count
top_year = year_counts.idxmax()
top_count = year_counts.max()

print(f"The year with the most 'Action' or 'Romance' movies is: {int(top_year)} ({top_count} movies)")

# Add bar chart
plt.figure(figsize=(12, 6))
plt.bar(year_counts.index, year_counts.values, color='coral')
plt.xlabel('Release Year')
plt.ylabel('Number of Action or Romance Movies')
plt.title(" Number of Action or Romance Movies Released per Year")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Q8: Find the movies with top 5 revenue - budget value

#sheet already loaded 
#profit already calculated in Q6

# Sort the data by profit in descending order
sorted_by_profit = data.sort_values(by='profit', ascending=False).reset_index(drop=True)

# Select the top 5 movies with the highest profit
top_5_profitable_movies = sorted_by_profit[['title', 'budget', 'revenue', 'profit']].head(5)

# Display the result
print("Top 5 movies by profit (revenue - budget):")
print(top_5_profitable_movies)

#Visualize the top 5 profitable movies
plt.figure(figsize=(10, 6))
plt.barh(top_5_profitable_movies['title'][::-1], top_5_profitable_movies['profit'][::-1], color='green')
plt.xlabel("Profit (USD)")
plt.title("Top 5 Most Profitable Movies")
plt.tight_layout()
plt.show()
```

```python
# Q9: Find the year with the highest number of movies release

#sheet already loaded
#dates already converted to datatime 

# Count the number of movies released per year
movie_counts_per_year = data['release_year'].value_counts()

# Find the year with the most movies
top_year = int(movie_counts_per_year.idxmax())
top_count = movie_counts_per_year.max()

# Display result
print(f"The year with the highest number of movie releases is: {top_year} ({top_count} movies)")

# Plotting the result
plt.figure(figsize=(12, 6))
plt.bar(movie_counts_per_year.index, movie_counts_per_year.values, color='skyblue')
plt.xlabel("Release Year")
plt.ylabel("Number of Movies Released")
plt.title("Number of Movies Released Per Year")
plt.tight_layout()
plt.show()
```

```python
# Q10: Find the top 2 countries with the highest production movies number

#import Counter to be able to count
from collections import Counter

# Convert 'production_countries' from string to list of dictionaries
data['production_countries'] = data['production_countries'].apply(lambda x: eval(x) if isinstance(x, str) else [])

# Extract country names
data['country_names'] = data['production_countries'].apply(lambda x: [d['name'] for d in x if 'name' in d])

# Count all country names
country_counter = Counter()
for country_list in data['country_names']:
    country_counter.update(country_list)

# Display the top 2 countries
print("Top 2 countries with the most produced movies:")
for country, count in country_counter.most_common(2):
    print(f"{country}: {count} movies")

    # Visualize the top 10 countries
top_10 = country_counter.most_common(10)
countries, counts = zip(*top_10)

plt.figure(figsize=(10, 6))
plt.barh(countries[::-1], counts[::-1], color='mediumseagreen')
plt.xlabel("Number of Movies Produced")
plt.title(" Top 10 Movie-Producing Countries")
plt.tight_layout()
plt.show()
```

```python
# Q11: Find the top 1 company with the highest production movies number

# Parse the 'production_companies' column safely
data['production_companies'] = data['production_companies'].apply(lambda x: eval(x) if isinstance(x, str) else [])

# Extract company names
data['company_names'] = data['production_companies'].apply(lambda x: [d['name'] for d in x if 'name' in d])

# Count all companies
company_counter = Counter()
for company_list in data['company_names']:
    company_counter.update(company_list)

# Show the top 1 company
top_1 = company_counter.most_common(1)
print("Top production company by number of movies:")
for name, count in top_1:
    print(f"{name}: {count} movies")
```
